```python
import pandas as pd
```

```python
# Importing Data
zone_data = pd.read_csv('./data/taxi_zone_lookup.csv')
zone_data
```

```python
# Replacing NaN values with Unknown

string_cols = zone_data.select_dtypes(include=["object"]).columns

zone_data[string_cols] = zone_data[string_cols].fillna("Unknown")
```

```python
# stripping away white spaces for consistency 
def strip_white_space(row):
    for items in row:
        items.strip()
    return row

zone_data[string_cols].apply(strip_white_space, axis='columns')

zone_data
```

```python
# Detecting duplicates

from collections import defaultdict

def detect_duplicates_using_dict(data):

    unique_location = defaultdict(int)
    duplicate_location_id= defaultdict(list)
    
    for _, row in data.iterrows():
        if (row.Borough,row.Zone,row.service_zone) in unique_location:
            duplicate_location_id[unique_location[(row.Borough,row.Zone,row.service_zone)]].append(row.LocationID)
        else:
            unique_location[(row.Borough,row.Zone,row.service_zone)] = row.LocationID
             
    return duplicate_location_id
            
    
detect_duplicates_using_dict(zone_data)


# duplicates using other cols except LocationId 

# data_with_req_attributes = data[['Borough','Zone','service_zone']]

# def find_duplicates(data):
#     duplicates = data[data.duplicated(keep=False)]
#     return duplicates
    
# find_duplicates(data_with_req_attributes)
```
